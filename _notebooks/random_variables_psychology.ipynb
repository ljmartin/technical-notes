# Random variables - the normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


fig, ax = plt.subplots(nrows=2, ncols=1)
fig.set_figheight(10)
fig.set_figwidth(10)

#plot a normal distribution probability density function
x = np.linspace(-5,5,100)
ax[0].plot(x, norm(0,1).pdf(x))
ax[0].axvline(0, c='k', linestyle='--')
ax[0].set_xlim(-5,5)

#plot a histogram of values drawn from the same normal distribution:
values = norm.rvs(0,1,100)
_ = ax[1].hist(values, bins=50)
ax[1].axvline(values.mean(), c='k', linestyle='--')
ax[1].set_xlim(-5,5)
#kdeplot(values, ax=ax[1])
```

- The curve in the top panel plots the relative probability of values
- If we simulate some samples (1000) we have estimated some properties of the random variable. Note that even at 100 samples the mean is off - we will come back to that later when we talk about standard error of the (estimated) mean.

# Properties of a normal distribution - i.e. the mean ($\mu$) and variance ($\sigma$)


A normal distribution can be characterised by two numbers (a.k.a. the parameters), which together describe:

* where the normal distribution is centred (the mean parameter - called $\mu$, pronounced `mew`), and
* how wide the normal distribution is (the variance parameter - called $\sigma$, pronounced `sigma`) 

The full equation that gives the relative probbailities of values, and thus describes the normal distribution, is a mess and we won't spend much time on it:

\begin{align}
p(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{\frac{-(x-\mu)^2}{2 \sigma^2}}
\end{align}

which in English says "the probability of value x, $p(x)$,  is given by... " So if you have the mean ($\mu$) and variance ($\sigma$) We can simplify it a lot to help understand it. This whole equation has the form 

\begin{align}
p(x) =  e^{-(x)^2}
\end{align}

Since $e$ is just a number that is around $2.7$, this is just:

\begin{align}
p(x) =  e^{-(x)^2}
\end{align}
